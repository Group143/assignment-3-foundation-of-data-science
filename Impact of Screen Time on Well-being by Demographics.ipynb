```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
demographic_data = pd.read_csv('/dataset1.csv')
screen_time_data = pd.read_csv('/dataset2.csv')
well_being_data = pd.read_csv('/dataset3.csv')

# Merge datasets on ID
merged_data = pd.merge(screen_time_data, demographic_data, on='ID')
merged_data = pd.merge(merged_data, well_being_data, on='ID')

# Define screen time variables
screen_time_vars = ['C_we', 'C_wk', 'G_we', 'G_wk', 'S_we', 'S_wk', 'T_we', 'T_wk']

# Function to analyze impact of screen time on well-being by demographic groups
def analyze_impact(demo_var):
    for var in screen_time_vars:
        print(f"\nAnalyzing impact of {var} on well-being by {demo_var}")
        # Group by demographic variable and screen time variable
        group_means = merged_data.groupby(demo_var).agg({var: 'mean', 'Optm': 'mean', 'Usef': 'mean'}).reset_index()
        print(group_means)

        # Plot average well-being indicators vs. screen time by demographic group
        plt.figure(figsize=(12, 6))

        # Plot for Optm
        plt.subplot(1, 2, 1)
        sns.barplot(x=demo_var, y='Optm', data=group_means)
        plt.title(f'Average Optimism by {var} and {demo_var}')
        plt.xlabel(demo_var)
        plt.ylabel('Average Optimism')

        # Plot for Usef
        plt.subplot(1, 2, 2)
        sns.barplot(x=demo_var, y='Usef', data=group_means)
        plt.title(f'Average Feeling Useful by {var} and {demo_var}')
        plt.xlabel(demo_var)
        plt.ylabel('Average Feeling Useful')

        plt.tight_layout()
        plt.show()

# Analyze impact by gender
analyze_impact('gender')

# Analyze impact by minority status
analyze_impact('minority')

# Analyze impact by deprivation status
analyze_impact('deprived')

# Scatter plots of screen time vs well-being by demographic groups
def scatter_plots_by_demo(demo_var):
    plt.figure(figsize=(20, 15))
    for i, var in enumerate(screen_time_vars):
        plt.subplot(4, 4, i + 1)
        sns.scatterplot(x=merged_data[var], y=merged_data['Optm'], hue=merged_data[demo_var], palette='viridis')
        plt.title(f'{var} vs Optimism by {demo_var}')
        plt.xlabel(var)
        plt.ylabel('Optimism')

        plt.subplot(4, 4, i + 9)
        sns.scatterplot(x=merged_data[var], y=merged_data['Usef'], hue=merged_data[demo_var], palette='viridis')
        plt.title(f'{var} vs Feeling Useful by {demo_var}')
        plt.xlabel(var)
        plt.ylabel('Feeling Useful')

    plt.tight_layout()
    plt.show()

# Plot scatter plots for gender
scatter_plots_by_demo('gender')

# Plot scatter plots for minority status
scatter_plots_by_demo('minority')

# Plot scatter plots for deprivation status
scatter_plots_by_demo('deprived')
```
