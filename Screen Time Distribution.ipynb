```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/dataset2.csv')

# Inspect the first few rows of the dataset
print(data.head())

# Check for missing values and data types
print(data.info())

# Summary statistics for each screen time variable
screen_time_vars = ['C_we', 'C_wk', 'G_we', 'G_wk', 'S_we', 'S_wk', 'T_we', 'T_wk']
print(data[screen_time_vars].describe())

# Set up the plotting environment
plt.figure(figsize=(16, 12))

# Plot histograms for each screen time variable
for i, var in enumerate(screen_time_vars, 1):
    plt.subplot(4, 2, i)
    sns.histplot(data[var], bins=30, kde=True)
    plt.title(f'Histogram of {var}')
    plt.xlabel('Hours per Day')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Plot box plots for each screen time variable
plt.figure(figsize=(16, 12))
for i, var in enumerate(screen_time_vars, 1):
    plt.subplot(4, 2, i)
    sns.boxplot(x=data[var])
    plt.title(f'Box Plot of {var}')
    plt.xlabel('Hours per Day')

plt.tight_layout()
plt.show()
```
